```python
from openai import OpenAI
import requests
from bs4 import BeautifulSoup

client = OpenAI()
```

```python
AGENT_GOAL = """
Read a public webpage and produce a clear 3-bullet-point summary
for a non-technical audience.
"""
```

```python
planning_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are an AI agent that plans steps to achieve a goal."},
        {"role": "user", "content": f"Goal:\n{AGENT_GOAL}\n\nCreate step-by-step plan."}
    ]
)

plan = planning_response.choices[0].message.content
print("AGENT PLAN:\n", plan)
```

```python
url = "https://en.wikipedia.org/wiki/Generative_artificial_intelligence"

html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
text = soup.get_text(separator=" ", strip=True)

# Limit content for safety
text = text[:3000]
```

```python
summary_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are an AI agent producing executive summaries."},
        {"role": "user", "content": f"Summarize the following content in 3 bullet points:\n{text}"}
    ]
)

print("FINAL OUTPUT:\n", summary_response.choices[0].message.content)
```
